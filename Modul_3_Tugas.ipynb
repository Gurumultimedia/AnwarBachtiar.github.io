### Soal 3 (nilai 50)

Lakukan looping di mana kita bisa mencetak semua bilangan prima, sekaligus bisa memunculkan:<br>
1.   Hasil penjumlahan semua bilangan prima yang muncul
2.   Hasil kuadrat semua bilangan prima yang muncul dalam bentuk list
3.   Hasil akar semua bilangan prima yang muncul dalam bentuk list dalam format 2 angka di belakang koma
4.   Hasil penjumlahan dari semua nilai di poin 2 dan 3 dalam bentuk list

Hasil eksekusinya tampak sebagai berikut:<br>

<img src="https://www.megabagus.id/wp-content/uploads/2024/07/prima.jpg" alt="prima" border="0">

```python
# Tulis program Python di sini
import math

# Input nilai awal dan akhir
awal = int(input("Masukkan nilai awal = "))
akhir = int(input("Masukkan nilai akhir = "))

bilangan_prima = []

# Mencari bilangan prima
for angka in range(awal, akhir + 1):
    if angka > 1:
        prima = True
        for i in range(2, angka):
            if angka % i == 0:
                prima = False
                break
        if prima:
            bilangan_prima.append(angka)

# 1. Hasil penjumlahan bilangan prima
jumlah_prima = sum(bilangan_prima)

# 2. Hasil kuadrat bilangan prima
kuadrat_prima = [x**2 for x in bilangan_prima]

# 3. Hasil akar bilangan prima (2 angka di belakang koma)
akar_prima = [round(math.sqrt(x), 2) for x in bilangan_prima]

# 4. Hasil penjumlahan poin 2 dan 3
jumlah_kuadrat_akar = [round(kuadrat_prima[i] + akar_prima[i], 2) for i in range(len(bilangan_prima))]

# Output
print("Bilangan prima =", bilangan_prima)
print("1. Hasil penjumlahan bilangan prima =", jumlah_prima)
print("2. Hasil kuadrat bilangan prima =", kuadrat_prima)
print("3. Hasil akar bilangan prima =", akar_prima)
print("4. Hasil penjumlahan poin 2 dan 3 =", jumlah_kuadrat_akar)
```

```python
import math

# Input dari user
awal = int(input("Masukkan nilai awal = "))
akhir = int(input("Masukkan nilai akhir = "))

prima = []

# Mencari bilangan prima
for x in range(awal, akhir + 1):
    if x > 1:
        is_prima = True
        for i in range(2, x):
            if x % i == 0:
                is_prima = False
                break
        if is_prima:
            prima.append(x)

# 1. Penjumlahan bilangan prima
jumlah_prima = sum(prima)

# 2. Kuadrat bilangan prima
kuadrat = [i**2 for i in prima]

# 3. Akar bilangan prima (2 angka di belakang koma)
akar = [round(math.sqrt(i), 2) for i in prima]

# 4. Penjumlahan poin 2 dan 3
jumlah_kuadrat_akar = [round(kuadrat[i] + akar[i], 2) for i in range(len(prima))]

# Output
print("Bilangan prima =", prima)
print("1. Hasil penjumlahan bilangan prima =", jumlah_prima)
print("2. Hasil kuadrat bilangan prima =", kuadrat)
print("3. Hasil akar bilangan prima =", akar)
print("4. Hasil penjumlahan poin 2 dan 3 =", jumlah_kuadrat_akar)
```
