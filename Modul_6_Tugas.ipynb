```python
!pip install pandas
import pandas as pd
import numpy as np
df = pd.read_csv('/content/Latihan_preprocessing_lv2.csv')
df
```

```python
#Lihat Struktur Awal
df.info()
df.head()
df.columns = df.columns.str.strip()
```

```python
# Bersihkan Nama Kolom, Hilangkan spasi dan rapikan penamaan
df.columns = df.columns.str.strip()
df
```

```python
# Tangani Missing Value, Dari gambar, kolom “Lainnya” menggunakan nilai "Kosong"
df['Other_Social_Network'] = df['Other_Social_Network'].fillna('Kosong')

df.fillna({
    'Preferred_Email': 'Unknown',
    'Preferred_Browser': 'Unknown'
}, inplace=True)

print(df.head())
```

```python
df
```

```python
# Ambil semua kolom kategorik/object .mode()
kolom_kategorik = df.select_dtypes(include='object').columns

# Isi kolom kosong  dengan modus
for col in kolom_kategorik:
    modus = df[col].mode()[0]
    df[col] = df[col].fillna(modus)

# Cek hasil
print(df.isnull().sum())
```

```python
df
```

```python
# Seragamkan penulisan sosial media
df['Other_Social_Network'] = df['Other_Social_Network'].replace({
    'IG': 'Instagram',
    'Insta': 'Instagram',
    'instagram': 'Instagram'
})

# Lihat hasil
print(df[['Other_Social_Network']].head(10))
```

```python
# Kolom kategorik
col = 'Other_Social_Network'

# Jadikan angka menjadi NaN
df[col] = df[col].replace(r'^\d+$', np.nan, regex=True)

# Isi NaN dengan modus
modus = df[col].mode()[0]
df[col] = df[col].fillna(modus)

# Lihat hasil
print(df[col].unique())
df
```

```python
# Ganti nama kolom dengan bahasa Indonesia
df = df.rename(columns={
    'Gender': 'Kelamin',
    'Race': 'Ras',
    'Birth_Year': 'TTL',
    'Marital_Status': 'Status',
    'Years_on_Internet': 'Durasi',
    'Hours_Per_Day': 'Jam',
    'Preferred_Browser': 'Browser',
    'Preferred_Search_Engine': 'Search',
    'Preferred_Email': 'Email',
    'Read_News': 'News',
    'Online_Shopping': 'Shopping',
    'Online_Gaming': 'Gaming',
    'Other_Social_Network': 'Lainnya'
})

#Lihat Struktur Akhir
df.info()
df.head()
```

```python
kolom_yn = ['News', 'Shopping', 'Gaming', 'Facebook', 'Twitter']

for kolom in kolom_yn:

    # Ambil data valid
    data_valid = df[df[kolom].isin(['Y', 'N'])][kolom]

    # Cari modus
    modus = data_valid.mode()[0]

    # Ganti selain Y/N dengan modus
    df.loc[~df[kolom].isin(['Y', 'N']), kolom] = modus
```

```python
# Simpan hasil data cleaning ke file CSV baru
df.to_csv('hasil_preprocessing.csv', index=False)

print("File berhasil disimpan")


# Memastikan file tersimpan
import os
print(os.listdir())
df
```
